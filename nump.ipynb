```python
import matplotlib.pyplot as plt
import numpy as np 
x = np.linspace(0,5,11)
y = x**2
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
#range from 0-1 i.e the dimension of graph
axes.plot(x,y,'b')
axes.set_xlabel('Intrest reciprocated')
axes.set_ylabel('Lpu')
axes.set_title('Truth')
```
