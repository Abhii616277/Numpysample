```python
import numpy as np
Marks=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
print(Marks)
```

```python
print(Marks.size)
print(Marks.shape)
print(Marks.dtype)

print(Marks[1])
print(Marks[:, 2])
np.sum(Marks+5)
```

```python
# for  row in Marks:
#     for i in row:
#         if i>= 50:
#             print("Pass")
#         else:
#             print("Fail")
result  = np.where(Marks>50,"Pass","Fail")
print(result)
```

```python
average = np.mean(Marks,axis=0)
result = np.where(average>=85,"GRADE A ",np.where(average>=70,"GRADE B",np.where(average>=50,"GRADE c","GRADE D")))
print(result)
```
